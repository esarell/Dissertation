##Sentence Level


Use this on the results outputed by Mengs Experimental MT System. So MMT and NV-MT


```python
import json
import csv
import torch
from transquest.algo.sentence_level.monotransquest.run_model import MonoTransQuestModel
model = MonoTransQuestModel("xlmroberta", "TransQuest/monotransquest-da-en_zh-wiki", num_labels=1, use_cuda=torch.cuda.is_available())
cvf = open('QESentenceValues.csv', 'w')
f = open("QEdata.json", "r")
jsonfile = json.load(f)
writer = csv.writer(cvf)
data =[]
for i in jsonfile:
  data.append([i["decoded_sequence"],i["input_sequence"]])

predictions, raw_outputs = model.predict(data)
writer.writerow(predictions)
cvf.close()
f.close() 
```

## Word Level

```python
import json
import csv
from transquest.algo.word_level.microtransquest.run_model import MicroTransQuestModel
import torch


model = MicroTransQuestModel("xlmroberta", "TransQuest/microtransquest-en_zh-wiki", labels=["OK","BAD"], use_cuda=torch.cuda.is_available())

cvf = open('QESourceWordValues.csv', 'w')
cvt = open('QETargetWordValues.csv', 'w')
f = open("QEdata.json", "r")
jsonfile = json.load(f)
firstwriter = csv.writer(cvf)
secondwriter=csv.writer(cvt)
data =[]
for i in jsonfile:
  data.append([i["decoded_sequence"],i["input_sequence"]])

source_tags, target_tags =  model.predict(data)

firstwriter.writerow(source_tags)
secondwriter.writerow(target_tags)
cvt.close()
cvf.close()
f.close() 
```
